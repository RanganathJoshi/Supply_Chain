```python
def detect_outliers(col):
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    upper_range=Q3+(IQR*1.5)
    lower_range=Q1-(IQR*1.5)
    return lower_range,upper_range
```

```python
col='unit_of_measure_(per_pack)'
l,u=detect_outliers(df[col])
l,u
```

### Backward Elimination using OLS method

```python
cols=list(x_train_data.columns)
pmax=1
while (len(cols)>0):
    x_1=x_train_data[cols]
    x_1=smf.add_constant(x_1)
    lr=smf.OLS(y_train,x_1).fit()
    p=pd.Series(lr.pvalues.values[1:],index=cols)
    pmax=max(p)
    feature_with_high_p=p.idxmax()
    if pmax>0.05:
        cols.remove(feature_with_high_p)
    else:
        break
selected_cols=cols
print(selected_cols)
    
    
```
